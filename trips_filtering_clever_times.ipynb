```python
details = pd.DataFrame()
c = 1
for trip in trips:
  details.loc[c, 'tripd_id'] = trip.origtatripno.iloc[0]
  details.loc[c, 'clever_start_time'] = trip.tmstmp.iloc[0]
  details.loc[c, 'clever_start_timestamp'] = trip.timestamp.iloc[0]
  details.loc[c, 'clever_end_time'] = trip.tmstmp.iloc[-1]
  details.loc[c, 'clever_end_timestamp'] = trip.timestamp.iloc[-1]
  c+=1
```

```python
details.sort_values(['clever_start_timestamp']).reset_index().drop(['index'], axis=1)[3:].to_csv('20220111_clever_times.csv')
```
