```python
import json
with open("bustime_chattanooga_jan_2022.json", "r") as f:
    data = json.load(f)

dated = []
for record in data:
    tmstmp = record["tmstmp"]
    if "20220111" in tmstmp:

        dated.append(record)
   

```

```python
data =dated
dated = []
for record in data:
    trp = record["origtatripno"]
    if trp:
        dated.append(trp)

unique_trp = list(set(dated))
unique_trp.sort()

print(len(unique_trp))
```

```python
import pandas as pd
import json
df1 = pd.read_csv("./20211024/trips.txt", sep=",")
df2 = pd.read_csv("./20211024/stop_times.txt", sep=",")

trips = unique_trp
trips = list(map(int, trips))

df1_fil  = pd.DataFrame(columns = df1.columns)
df2_fil  = pd.DataFrame(columns = df2.columns)

indices = []
for trip in trips:
    for index, row in df1.iterrows():       
        if trip == int(row['trip_id']/1000):
            
            indices.append(index)
            df1_fil.append(row, ignore_index=True)
            break

df1_fil = df1.iloc[indices]

indices = []
for trip in trips:
    for index, row in df2.iterrows():       
        if trip == int(row['trip_id']/1000):
            indices.append(index)
            df2_fil.append(row, ignore_index=True)

# df2_fil = df2.iloc[indices]
# print(df1_fil.head())
# df_fil.to_csv('df_fil_19.csv', index=False)
```
